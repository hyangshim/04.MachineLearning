```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
df_2004 =pd.read_csv('data/건강검진정보_2004.CSV',encoding='euc-kr')
df_2004.head(3)
```

```python
df_2009 =pd.read_csv('data/건강검진정보_2009(수정).csv',encoding='euc-kr')
df_2009.head(3)
```

```python
df_2014 =pd.read_csv('data/건강검진정보_2014.CSV',encoding='euc-kr')
df_2014.head(3)
```

```python
df_2019 =pd.read_csv('data/건강검진정보_2019(수정).csv',encoding='euc-kr')
df_2019=df_2019.iloc[:1000000,:]
df_2019.head(3)
```

```python
df_2004=df_2004[['연령대코드(5세단위)','성별코드','(혈청지오티)AST','(혈청지오티)ALT','감마지티피']]
df_2004.dropna(inplace=True)
df_2004.isna().sum()
```

```python
df_2009=df_2009[['연령대 코드(5세 단위)','성별코드','(혈청지오티) AST','(혈청지오티)ALT','감마 지티피']]
df_2009.columns =['연령대코드(5세단위)','성별코드','(혈청지오티)AST','(혈청지오티)ALT','감마지티피']
df_2004.dropna(inplace=True)
df_2004.isna().sum()
```

```python
df_2014=df_2014[['연령대코드(5세단위)','성별코드','(혈청지오티)AST','(혈청지오티)ALT','감마지티피']]
df_2014.columns =['연령대코드(5세단위)','성별코드','(혈청지오티)AST','(혈청지오티)ALT','감마지티피']
df_2004.dropna(inplace=True)
df_2004.isna().sum()
```

```python
df_2019=df_2019[['연령대 코드(5세단위)','성별코드','(혈청지오티)AST','(혈청지오티)ALT','감마 지티피']]
df_2019.columns =['연령대코드(5세단위)','성별코드','(혈청지오티)AST','(혈청지오티)ALT','감마지티피']
df_2019.dropna(inplace=True)
df_2019.isna().sum()
```

```python
#남자 여자 으로 분리
df_2019_남자=df_2019[df_2019.성별코드 == 1]
df_2019_여자=df_2019[df_2019.성별코드 == 2]
```

```python
df_2019_남자_선별=df_2019_남자.loc[(df_2019_남자['(혈청지오티)AST']>40) | (df_2019_남자['(혈청지오티)ALT']>40)|(df_2019_남자['감마지티피']>63)]
df_2019_여자_선별=df_2019_여자.loc[(df_2019_여자['(혈청지오티)AST']>40) | (df_2019_여자['(혈청지오티)ALT']>40)|(df_2019_여자['감마지티피']>35)]
df_2019_남자_선별.head(3)
```

```python
#남자 카운트으로 바꿈
df_2019_count_남자=df_2019_남자_선별.groupby(['연령대코드(5세단위)','성별코드'])[['(혈청지오티)AST']].count()
df_2019_count_남자=pd.DataFrame(df_2019_count_남자)
total=pd.DataFrame(df_2019_남자.groupby(['연령대코드(5세단위)', '성별코드'])['(혈청지오티)AST'].count())
```

```python
# 2019년 남자 간질환 유병률
df_liver_2019_남자=(df_2019_count_남자/total*100).round(1)
df_liver_2019_남자
df_liver_2019_남자.reset_index(inplace=True)
df_liver_2019_남자.head(3)
```

```python
df_liver_2019_남자.set_index('연령대코드(5세단위)',inplace=True) 
```

```python
#여자 카운트으로 바꿈
df_2019_count_여자=df_2019_여자_선별.groupby(['연령대코드(5세단위)','성별코드'])[['(혈청지오티)AST']].count()
df_2019_count=pd.DataFrame(df_2019_count_여자)
total=pd.DataFrame(df_2019_여자.groupby(['연령대코드(5세단위)', '성별코드'])['(혈청지오티)AST'].count())
```

```python
# 2019년 여자 간질환 유병률
df_liver_2019_여자=(df_2019_count_여자/total*100).round(1)
df_liver_2019_여자.reset_index(inplace=True)
df_liver_2019_여자.tail(3)
```

```python
df_liver_2019_여자.set_index('연령대코드(5세단위)',inplace=True)
```

```python
plt.bar(df_liver_2019_여자.index,df_liver_2019_남자['(혈청지오티)AST'], label='남자',width=0.4)
plt.bar(df_liver_2019_여자.index+0.4,df_liver_2019_여자['(혈청지오티)AST'],  label='여자',width=0.4)
plt.xticks(np.arange(5, 19, 1))
plt.legend()
plt.ylim(0,50)
plt.title('2019년 연령별/성별 간질환 유병률')
plt.show()
```

```python
df_liver_heat_2=df_liver_2019_여자.reset_index().pivot(index='연령대코드(5세단위)',columns='성별코드',values='(혈청지오티)AST')
df_liver_heat_1=df_liver_2019_남자.reset_index().pivot(index='연령대코드(5세단위)',columns='성별코드',values='(혈청지오티)AST')
df_liver_heat_1.head()
```

```python
#남자컬럼에 여자 컬럼 추가
df_liver_heat_1[2]=df_liver_heat_2[2]
df_liver_heat_1
```

```python
import seaborn as sns
sns.heatmap(df_liver_heat_1.sort_index(ascending=False),annot=True,xticklabels=['남자','여자'])
plt.title('2019년 연령별/성별 간질환 유병률');
```
