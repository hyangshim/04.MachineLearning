### 2009년~2019년 성별 이상지질혈증 유병률 변화 

```python
df_2009_1=df_2009.loc[(df_2009['총 콜레스테롤']>=201) | (df_2009['트리글리세라이드']>=151)|(df_2009['HDL 콜레스테롤']>=61)|(df_2009['LDL 콜레스테롤']>=131)]
df_2014_1=df_2014.loc[(df_2014['총콜레스테롤']>=201) | (df_2014['트리글리세라이드']>=151)|(df_2014['HDL콜레스테롤']>=61)|(df_2014['LDL콜레스테롤']>=131)]
df_2019_1=df_2019.loc[(df_2019['총 콜레스테롤']>=201) | (df_2019['트리글리세라이드']>=151)|(df_2019['HDL 콜레스테롤']>=61)|(df_2019['LDL 콜레스테롤']>=131)]
```

```python
df_2009_count=df_2009_1.groupby(['연령대 코드(5세 단위)','성별코드'])[['총 콜레스테롤']].count().round(1)
df_2009_count=pd.DataFrame(df_2009_count)
df_2014_count=df_2014_1.groupby(['연령대코드(5세단위)','성별코드'])[['총콜레스테롤']].count().round(1)
df_2014_count=pd.DataFrame(df_2014_count)
df_2019_count=df_2019_1.groupby(['연령대 코드(5세단위)','성별코드'])[['총 콜레스테롤']].count().round(1)
df_2019_count=pd.DataFrame(df_2019_count)
```

```python
total_2009=pd.DataFrame(df_2009.groupby(['연령대 코드(5세 단위)', '성별코드'])['총 콜레스테롤'].count())
total_2014=pd.DataFrame(df_2014.groupby(['연령대코드(5세단위)', '성별코드'])['총콜레스테롤'].count())
total_2019=pd.DataFrame(df_2019.groupby(['연령대 코드(5세단위)', '성별코드'])['총 콜레스테롤'].count())
```

```python
df_hemostasis_2009=(df_2009_count/total_2009*100).round(1)
df_hemostasis_2014=(df_2014_count/total_2014*100).round(1)
df_hemostasis_2019=(df_2019_count/total_2019*100).round(1)
df_hemostasis_2009.head()
```

```python
df_final_2009=pd.DataFrame(df_hemostasis_2009.groupby(['성별코드'])['총 콜레스테롤'].mean().round(1))
df_final_2014=pd.DataFrame(df_hemostasis_2014.groupby(['성별코드'])['총콜레스테롤'].mean().round(1))
df_final_2019=pd.DataFrame(df_hemostasis_2019.groupby(['성별코드'])['총 콜레스테롤'].mean().round(1))
df_final_2009.head()
```

```python
df_last = pd.concat([df_final_2009,df_final_2014,df_final_2019], axis=1)
df_last.columns = ['2009','2014','2019']
df_last.index=['남자','여자']
df_last
```

```python
df_last.T.plot(marker='8')
plt.title('2009년~2019년 성별 이상지질혈증 유병률 변화')
plt.xticks([0,1,2])
plt.ylim(50,80)
plt.show()
```
