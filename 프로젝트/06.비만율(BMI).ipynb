```python
# 2004년~2019년 BMI 추가함
bmi_2004 =round(df_2004['체중(5Kg 단위)'] /(df_2004['신장(5Cm단위)']/100) **2,1)
bmi_2009 =round(df_2009['체중(5Kg 단위)'] /(df_2009['신장(5Cm단위)']/100) **2,1)
bmi_2014 =round(df_2014['체중(5Kg 단위)'] /(df_2014['신장(5Cm단위)']/100) **2,1)
bmi_2019 =round(df_2019['체중(5Kg 단위)'] /(df_2019['신장(5Cm단위)']/100) **2,1)
```

```python
df_2004['BMI'] = bmi_2004
df_2009['BMI'] = bmi_2009
df_2014['BMI'] = bmi_2014
df_2019['BMI'] = bmi_2019
```

```python
df_2004_1=df_2004.loc[(df_2004['BMI']>25)]
df_2009_1=df_2009.loc[(df_2009['BMI']>25)]
df_2014_1=df_2014.loc[(df_2014['BMI']>25)]
df_2019_1=df_2019.loc[(df_2019['BMI']>25)]
```

```python
df_2004_count=df_2004_1.groupby(['연령대코드(5세단위)','성별코드'])[['BMI']].count().round(1)
df_2004_count=pd.DataFrame(df_2004_count)
df_2009_count=df_2009_1.groupby(['연령대 코드(5세 단위)','성별코드'])[['BMI']].count().round(1)
df_2009_count=pd.DataFrame(df_2009_count)
df_2014_count=df_2014_1.groupby(['연령대코드(5세단위)','성별코드'])[['BMI']].count().round(1)
df_2014_count=pd.DataFrame(df_2014_count)
df_2019_count=df_2019_1.groupby(['연령대 코드(5세단위)','성별코드'])[['BMI']].count().round(1)
df_2019_count=pd.DataFrame(df_2019_count)
```

```python
total_2004=pd.DataFrame(df_2004.groupby(['연령대코드(5세단위)', '성별코드'])['BMI'].count())
total_2009=pd.DataFrame(df_2009.groupby(['연령대 코드(5세 단위)', '성별코드'])['BMI'].count())
total_2014=pd.DataFrame(df_2014.groupby(['연령대코드(5세단위)', '성별코드'])['BMI'].count())
total_2019=pd.DataFrame(df_2019.groupby(['연령대 코드(5세단위)', '성별코드'])['BMI'].count())
```

```python
df_hemostasis_2004=(df_2004_count/total_2004*100).round(1)
df_hemostasis_2009=(df_2009_count/total_2009*100).round(1)
df_hemostasis_2014=(df_2014_count/total_2014*100).round(1)
df_hemostasis_2019=(df_2019_count/total_2019*100).round(1)
```

```python
df_final_2004=pd.DataFrame(df_hemostasis_2004.groupby(['성별코드'])['BMI'].mean().round(1))
df_final_2009=pd.DataFrame(df_hemostasis_2009.groupby(['성별코드'])['BMI'].mean().round(1))
df_final_2014=pd.DataFrame(df_hemostasis_2014.groupby(['성별코드'])['BMI'].mean().round(1))
df_final_2019=pd.DataFrame(df_hemostasis_2019.groupby(['성별코드'])['BMI'].mean().round(1))
```

```python
df_last = pd.concat([df_final_2004,df_final_2009,df_final_2014,df_final_2019], axis=1)
df_last.columns = ['2004','2009','2014','2019']
df_last.index=['남자','여자']
df_last
```

```python
df_last.T.plot(marker='8')
plt.title('2004년~2019년 성별 비만율(BMI) 변화 ')
plt.xticks([0,1,2,3])
plt.ylim(10,50)
plt.show()
```
